```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from operator import itemgetter
```

```python
def yeild_plot_US(maturity1,maturity2):
    df1 = pd.read_csv("https://raw.githubusercontent.com/POTACCC/C4RM-Final-Project/main/daily-treasury-rates.csv")
    sns.heatmap(df1.corr(), annot = True, fmt='.2g',cmap= 'coolwarm')
    plt.show()
    df1['Month'] = df1['Date'].apply(lambda i: i.split('/')[0]).astype(int)
    plt1 = sns.lineplot(data=df1, x="Month", y=df1[maturity1],estimator='mean',err_style=None,color = 'blue')
    plt2 = sns.lineplot(data=df1, x="Month", y=df1[maturity2],estimator='mean',err_style=None,color = 'green')
    df1['Spreads']=df1[maturity2]-df1[maturity1]
    plt3 = sns.lineplot(data=df1, x="Month", y="Spreads",estimator='mean',err_style=None,color = 'red')
    plt.plot([1, len(np.unique(df1['Month']))], [0, 0], color = 'black')
    plt.title('US Par Yield Curve in 2022 for Maturity in ' + maturity1 + ' and '+ maturity2 + ' years')
    plt.xlabel('Month')
    plt.ylabel('Yield (%)')
    plt.legend([maturity1 + ' Year',maturity2+ ' Year','Spread'], ncol=2, loc='upper right')
    plt.show()
    i, value = min(enumerate(df1['Spreads']), key=itemgetter(1))
    Date= df1['Date'][i]
    print(f'The worst spread is {value:.2f}% on {Date}')
```

```python
yeild_plot_US("1","30")
```

```python
def yeild_plot_RU(maturity1,maturity2):
    df1 = pd.read_csv("https://raw.githubusercontent.com/POTACCC/C4RM-Final-Project/main/Yield%20Curve.csv")
    sns.heatmap(df1.corr(), annot = True, fmt='.2g',cmap= 'coolwarm')
    plt.show()
    df1['Month'] = df1['Date'].apply(lambda i: i.split('.')[1]).astype(int)
    plt1 = sns.lineplot(data=df1, x="Month", y=df1[maturity1],estimator='mean',err_style=None,color = 'blue')
    plt2 = sns.lineplot(data=df1, x="Month", y=df1[maturity2],estimator='mean',err_style=None,color = 'green')
    df1['Spreads']=df1[maturity2]-df1[maturity1]
    plt3 = sns.lineplot(data=df1, x="Month", y="Spreads",estimator='mean',err_style=None,color = 'red')
    plt.plot([1, len(np.unique(df1['Month']))], [0, 0], color = 'black')
    plt.title('Russian Government Bond Yield in 2022 for Maturity in ' + maturity1 + ' and '+ maturity2 + ' years')
    plt.xlabel('Month')
    plt.ylabel('Yield (%)')
    plt.legend([maturity1 + ' Year',maturity2+ ' Year','Spread'], ncol=2, loc='upper right')
    plt.show()
    i, value = min(enumerate(df1['Spreads']), key=itemgetter(1))
    Date= df1['Date'][i]
    print(f'The worst spread is {value:.2f}% on {Date}')
```

```python
yeild_plot_RU("1","30")
```
